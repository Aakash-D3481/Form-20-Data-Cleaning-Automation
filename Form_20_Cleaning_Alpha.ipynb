```python
import pandas as pd
pd.set_option("display.max_columns", 200)
```

```python
# Read the file
df = pd.read_excel(r"C:\Users\Aakash D\Downloads\Analytics\Self Paced Projects\Data Cleaning Automation\Raw- Don't Touch\401._Robertsganj.xls", header=None)

#Drop the redundant rows
df = df.drop(range(8)).reset_index(drop=True)
df = df.drop(columns=[3, 8], axis= 1)
df.head(10)
```

```python
#df.drop(columns= [0,1], axis=1, inplace=True
start_part = df.iloc[:,:8]
start_part

mid_part = df.iloc[:,8:50]
mid_part = mid_part.iloc[:,2::3]

end_part = df.iloc[:,-2:]
end_part

# renaming all the columns to Originals
```

```python
parties = ["CPI","BJP","SP","INC","BSP","QED","AIPF","BMP","AAP","SSP","CPI-M","IND","IND","IND"]
alpha = ["S.No.", "P.S.No.", "Constituency", "Male", "Female", "Others", "Total", "No. of Ten-dered Voters"]
omega = ["NOTA", "Total_Valid_Votes"]

start_part.columns = alpha
mid_part.columns = parties
end_part.columns = omega

final_df = pd.concat([start_part,mid_part,end_part], axis=1)
final_df = final_df.drop(final_df.index[-2:])
df = final_df
df.columns
```

```python
ind_cols = df.filter(like="IND").columns

df["IND (Sum)"] = df[ind_cols].sum(axis = 1).astype(int)
df["IND (Total)"] = (df["IND (Sum)"]/len(ind_cols)).astype(int)
df.drop(columns=["IND (Sum)"], axis=1, inplace=True)

df["Constituency"] = "Robertganj"

df["Total Valid Votes"] = df.iloc[:, df.columns.isin(parties)].sum(axis = 1).astype(int)
df.drop(columns=["Total_Valid_Votes"], axis=1, inplace=True)

df["Rejected Votes"] = 0

df
```

```python
df.to_excel("401._Robertsganj_Processed.xls", index= False)
```
